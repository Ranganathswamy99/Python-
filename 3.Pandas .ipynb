## Pandas Tutorial
Pandas is an open source, BSD-licensed library providing high-performance, easy-to-use data structures and data analysis tools for the Python programming language.

What is Data Frames?
DataFrame is a 2-dimensional labeled data structure in Pandas, like a table in Excel or a SQL table.
-It contains rows and columns.
-Columns can have different data types (int, float, string, etc.).
-Created using pd.DataFrame().

# CSV

```python
from io import StringIO, BytesIO
```

```python
data = ('col1,col2,col3\n'
            'x,y,1\n'
            'a,b,2\n'
            'c,d,3')
```

```python
StringIO()
```

```python
pd.read_csv(StringIO(data))
```

```python
## Read from specific columns
df=pd.read_csv(StringIO(data), usecols=['col1','col3'])
df
```

```python
data1 = ('a,b,c,d\n'
        '1,2,3,4\n'
        '5,6,7,8\n'
        '9,10,11,12')
```

```python
df=pd.read_csv(StringIO(data1),dtype=object)
df
```

```python
df=pd.read_csv(StringIO(data1),dtype=int)
df
```

```python
df=pd.read_csv(StringIO(data1),dtype=float)
df
```

```python
df['a']
```

```python
df=pd.read_csv(StringIO(data1),dtype={'b':int, 'c':float, 'a':'Int64'})
df
```

```python
df['a'][1]
```

```python
type(df['a'][1])
```

```python
df.dtypes
```

```python
data2=('index,a,b,c\n'
       '4,apple,bat,5.7\n'
       '8,orange,cow,10')
```

```python
pd.read_csv(StringIO(data2))
```

```python
pd.read_csv(StringIO(data2),index_col=1)
```

```python
pd.read_csv(StringIO(data2),index_col=2)
```

```python
data3=('a,b,c\n'
       '4,apple,bat,\n'
       '8,orange,cow,')
```

```python
pd.read_csv(StringIO(data3))
```

```python
pd.read_csv(StringIO(data3),index_col=False)
```

```python
### Combinig use_columns and index_columns
data4=('a,b,c\n'
       '4,apple,bat,\n'
       '8,orange,cow,')
pd.read_csv(StringIO(data4))
```

```python
pd.read_csv(StringIO(data4),usecols=['b','c'],index_col=False)
```

```python
### Quoting and Escape Characters. Very useful in NLP

data5 = 'a,b\n"Hello,\\"Ranganath\\", Nice to see you",9'
```

```python
pd.read_csv(StringIO(data5))
```

```python
## we can escape the \\ useing escapechar
pd.read_csv(StringIO(data5), escapechar='\\')  
```
