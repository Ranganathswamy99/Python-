```python

# Compute the x and y coordinates for points on a sine curve 
x = np.arange(0, 4 * np.pi, 0.1) 
y = np.sin(x) 
plt.title("sine wave form") 

# Plot the points using matplotlib 
plt.plot(x, y, 'b') 
plt.show() 
```

```python
#Subplot()
# Compute the x and y coordinates for points on sine and cosine curves 
x = np.arange(0, 5 * np.pi, 0.1) 
y_sin = np.sin(x) 
y_cos = np.cos(x)  
   
# Set up a subplot grid that has height 2 and width 1, 
# and set the first such subplot as active. 
plt.subplot(2, 1, 1)
   
# Make the first plot 
plt.plot(x, y_sin,'g--') 
plt.title('Sine')  
   
# Set the second subplot as active, and make the second plot. 
plt.subplot(2, 1, 2) 
plt.plot(x, y_cos,'b--') 
plt.title('Cosine')  
   
# Show the figure. 
plt.show()
```

## Pie Chart

```python
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes =[350,300,200,150]
colors = ['Red', 'blue', 'yellow' , 'green']
explode = (0.2, 0, 0, 0) ##explode 1st slice

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
       autopct='%1.1f%%', shadow=False)

plt.axis('equal') # make a circle
plt.show()
```

```python
# Data: time spent on daily activities
labels = ['Sleep', 'Work', 'Exercise', 'Leisure', 'Others']
sizes = [8, 9, 2, 3, 2]  # Total 24 hours
colors = ['skyblue', 'lightgreen', 'orange', 'violet', 'gray']

# Create pie chart
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)

# Add title
plt.title("Daily Time Distribution")

# Equal aspect ratio ensures pie is drawn as a circle
plt.axis('equal')

plt.show()
```
