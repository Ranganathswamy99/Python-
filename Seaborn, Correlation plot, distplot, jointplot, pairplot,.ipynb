## Corealation with Heatmap
A corelation heatmap uses colored cells, typically in monochromatic scale, to show a 2D corelation matrix(table) b/w two disceate domensions or event types,
is is very important in Feature selection

```python
df[['total_bill', 'tip','size']].corr()
```

```python
sns.heatmap(df[['total_bill', 'tip','size']].corr())
```

## Univariate analysis

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind= 'hex')
```

```python
sns.jointplot(x = 'tip', y='total_bill', data = df, kind = 'reg')
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue='sex')
```

## Distplot
Dist plot helps us to check the distribution of the columns feature

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```

## Categorical Plots
### seaborn also helps us in doing the analysis on Categorical Data points. in this section we will discuss about
    -boxplot
    -violinplot
    -countplot
    -bar plot

```python
## Count plot

sns.countplot(x='sex',data=df)
```

```python
##count plot
sns.countplot(x='day',data=df)
```

```python
sns.countplot(x='smoker',data=df)
```

```python
## Barplot
sns.barplot(x='total_bill',y='sex',data=df)
```

```python
sns.barplot(y='total_bill',x='smoker',data=df)
```

# Box Plot
A box and wishker plot(sometimes called a boxplot) is a graph that presents information from a five-number summary

```python
sns.boxplot(x='sex',y='total_bill', data=df)
```

```python
sns.boxplot(x='day',y='total_bill', data=df, palette='rainbow')
```

```python
sns.boxplot(data=df,orient='v')
```

```python
# categorize my data based on some other categories
sns.boxplot(x='total_bill',y='day',hue='smoker',data=df)
```

## Violin Plot
Violin plot helps us to see both the distribution of data in terms of Kernel density estimation and the box plot

```python
sns.violinplot(x="total_bill", y="day", data=df,palette='rainbow')
```
